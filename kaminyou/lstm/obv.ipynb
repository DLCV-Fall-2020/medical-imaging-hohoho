```python
import pandas as pd
import numpy as np
```

```python
def maptolevel(img_id):
    img_prefix = img_id.split("_")[0]
    level = img_id.split("_")[1].split(".")[0]
    if len(level) == 1:
        return f"{img_prefix}_0{level}.jpg"
    else:
        return img_id
```

```python
df = pd.read_csv("lstm_kaggle_ori_from_resnet18_0.75001_l32again_0.45.csv")
df["curated_ID"] = list(map(maptolevel, df.ID.values))
df = df.sort_values(by=['curated_ID']).drop(columns=['curated_ID'])
df = df.reset_index(drop=True)
```

```python
df_ori = pd.read_csv("lstm_kaggle_ori_from_resnet18_0.75001.csv")
df_ori["curated_ID"] = list(map(maptolevel, df_ori.ID.values))
df_ori = df_ori.sort_values(by=['curated_ID']).drop(columns=['curated_ID'])
df_ori = df_ori.reset_index(drop=True)
```

```python
def pattern2pattern(df, pattern_from, pattern_to):
    group_list = []
    for name, group in df.groupby("dirname"):
        for subtype in ["ich", "ivh", "sah", "sdh", "edh"]:
            pattern = ''.join(map(str, group[subtype].values))
            while pattern_from in pattern:
                print(f"revise {name}'s {subtype}")
                pattern = pattern.replace(pattern_from, pattern_to)
            group[subtype] = list(pattern)
        group_list.append(group)
    revised_df = pd.concat(group_list).sort_index(ascending=True)
    return revised_df
```

```python
def pattern2patternonce(df, pattern_from, pattern_to):
    group_list = []
    for name, group in df.groupby("dirname"):
        for subtype in ["ich", "ivh", "sah", "sdh", "edh"]:
            pattern = ''.join(map(str, group[subtype].values))
            if "11111111111" in pattern:
                if pattern_from in pattern:
                    print(f"revise {name}'s {subtype}")
                pattern = pattern.replace(pattern_from, pattern_to)
            group[subtype] = list(pattern)
        group_list.append(group)
    revised_df = pd.concat(group_list).sort_index(ascending=True)
    return revised_df
```

```python
df_ori_revised = pattern2pattern(df_ori, "101", "111")
df_ori_revised_two = pattern2pattern(df_ori_revised, "00100", "00000")
```
